```python
import yfinance as yf
import matplotlib.pyplot as plt

# Download stock data
data = yf.download('TCS.NS', start='2022-01-01', end='2025-01-01')
data['Daily Return'] = data['Close'].pct_change()

plt.plot(data.index, data['Daily Return'])
plt.title("Daily Returns of TCS")
plt.xlabel("Date")
plt.ylabel("Daily Return")
plt.grid(True)
plt.show()

print("Average Return:", data['Daily Return'].mean())
print("Volatility:", data['Daily Return'].std())
```
